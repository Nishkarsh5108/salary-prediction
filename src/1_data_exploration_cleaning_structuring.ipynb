```python
import pandas as pd
from sklearn.model_selection import train_test_split

print("Trashing the fake Kaggle test set...")

# 1. Load the RAW validation data (the messy one with strings)
raw_valid = pd.read_csv("../data/valid.csv")

# 2. Slice it 50/50 like a pizza
new_valid, new_test = train_test_split(raw_valid, test_size=0.5, random_state=42)

# 3. Overwrite the old files so the master script finds them automatically
new_valid.to_csv("../data/valid.csv", index=False)
new_test.to_csv("../data/test.csv", index=False)

print("Split complete. You now have a real validation and test set.")
print(f"Valid shape: {new_valid.shape}")
print(f"Test shape:  {new_test.shape}")
```
